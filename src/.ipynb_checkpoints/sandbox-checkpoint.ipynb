# FREQ DOMAIN FEATS?

```python
from itertools import chain
import pywt
bidder = bidders_d[list(bots_id)[10]]
all_bid_ids = chain(*bidder.bids_by_auction.values())
user_bids = [Bid(**bids_df.loc[int(bid_id)]) 
             for bid_id in all_bid_ids]
```

```python
lvls = 3
n_chunks = 3
n_wav_feats = 56
n_b_thresh = 50
n_tb_thresh = 30

wav_feats = {bidder_id: np.zeros(n_wav_feats) 
             for bidder_id in bidders_d.keys()}
tic = time()
for ii, bidder in enumerate(bidders_d.values()):

    all_bid_ids = list(chain(*bidder.bids_by_auction.values()))
    if len(all_bid_ids) < n_b_thresh:
        continue
        
    user_bids = [Bid(**bids_df.loc[int(bid_id)]) 
                 for bid_id in all_bid_ids]

    t = np.array(sorted([b.time for b in user_bids]))
    step = 1e10
    bins = np.arange(t.min(), t.max()+1, step)
    tb, _ = np.histogram(t, bins=bins)
    
    if len(tb) < n_tb_thresh:
        continue

    chunk_step = len(tb)/4
    chunk_sz = len(tb)/2
    c_feats = []
    for c_start in (np.arange(3)*chunk_step):
        tb_chunk = tb[c_start:c_start+chunk_step]
        decs = pywt.wavedec(tb_chunk, 'db5', level=lvls)

        dec_fs = np.array([np.array([np.abs(dec).mean(), dec.std()]) for dec in decs])
        c_feats.append(dec_fs)

    feats = [f for f in c_feats]    
    feats.append(c_feats[1] - c_feats[0])
    feats.append(c_feats[2] - c_feats[1])
    feats.append(c_feats[1] - c_feats[0])
    feats.append(c_feats[2] - 2*c_feats[1] + c_feats[0])
    
    feats = np.array(feats).flatten()

    wav_feats[bidder.bidder_id] = feats
    
    print '\r %d / %d \t %g' % (ii+1, len(bidders_d), time()-tic),
toc = time() -tic
print toc
```

```python
# Iterate through auctions - get bid list
# Exclude first bid
# Per bidder per auction: mean & std
    # d[bidder_id] = [].append(mean & std)
# Per bidder final feat: mean(mean & std), std(mean&std)
    # for each d.item(): np.array([(mu, sig),... ]) . mean() & .std()

dt_d = {bidder_id: [] for bidder_id in bidders_d.keys()}
tic = time()

for ii, a in enumerate(auctions_d.values()):
    bids = [Bid(**bids_df.loc[int(bid_id)]) 
                 for bid_id in a.bids]
    t = np.array([b.time for b in bids])
    dt = t[1:] - t[:-1]
    bidder_ids = np.array([b.bidder_id for b in bids])[1:]
    
    for bidder_id in set(bidder_ids):
        bidder_dt = dt[bidder_ids==bidder_id]
        feat = [bidder_dt.mean(), bidder_dt.std()]
        dt_d[bidder_id].append(feat)

    print '\r %d / %d' % (ii+1, len(auctions_d)),
toc = time() -tic
print toc

```

```python
dt_feats_d = {}
for ii, (bidder_id, vals) in enumerate(dt_d.items()):
    if len(vals):
        v_arr = np.array(vals)
        agg_feat = np.concatenate([
                v_arr.mean(axis=0),
                v_arr.std(axis=0),
                v_arr.max(axis=0),
                v_arr.min(axis=0),
            ])
    else:
        agg_feat = np.zeros(8)
    dt_feats_d[bidder_id] = agg_feat
    print '\r %d / %d' % (ii+1, len(dt_d)),
```

# RELATIVE POSITION OF BIDS


```python
from collections import Counter

pos_feats = {bidder_id: [] 
             for bidder_id in bidders_d.keys()}
tic = time()

for ii, a in enumerate(auctions_d.values()):
    if len(a.bids) > 1:
               
        bids = [Bid(**bids_df.loc[int(bid_id)]) 
                    for bid_id in a.bids]
        
        bidders_ids = np.array([b.bidder_id for b in bids])
        
        temp_d = {bidder_id: np.zeros(7)
             for bidder_id in set(bidders_ids)}
        
        t = np.array([b.time for b in bids])
        
        step = np.ceil((t.max()-t.min())/4.)
        tb = np.concatenate([np.arange(t.min(), t.max(), step), [t.max()+1]])
        tr = np.array([tb[:-1], tb[1:]]).T
        tr = np.r_[tr, tr[:3,:] + np.array([np.zeros(3), np.ones(3)*step]).T]
        
        for c_i, (c_start, c_end) in enumerate(tr):
            ids_in_c = bidders_ids[((t >= c_start) & (t < c_end))]
            count_d = Counter(ids_in_c)
            for k, v in count_d.items():
                add_v = np.zeros(7)
                add_v[c_i] = v
                temp_d[k] += add_v
        
        for bidder_id, cs in temp_d.items():
            pos_feats[bidder_id].append(cs)

    print '\r %d / %d \t %g' % (ii+1, len(auctions_d), time()-tic),
toc = time() -tic
print toc

```
